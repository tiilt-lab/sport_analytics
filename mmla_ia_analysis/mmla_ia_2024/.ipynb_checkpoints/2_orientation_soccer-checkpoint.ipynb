```python
key_joint = pd.read_csv("soccer_ep4.csv", index_col=0)
```

```python
key_joint
```

```python
key_joint['neck_estimate_x'] = np.average(key_joint[['left_shoulder_x', 'right_shoulder_x']], axis=1)
key_joint['neck_estimate_y'] = np.average(key_joint[['left_shoulder_y', 'right_shoulder_y']], axis=1)
```

```python
plt.hist(key_joint['neck_estimate_x'], color='steelblue', label='xvalues', alpha=0.5)
plt.hist(key_joint['neck_estimate_y'], color='orange', label='yvalues', alpha=0.5);
plt.legend();
```

```python
key_joint.shape
```

```python
# Body points
plt.scatter(-key_joint.iloc[258, 1:34], -key_joint.iloc[258, 34:67])
# Nose point
plt.scatter(-key_joint.iloc[258, 1], -key_joint.iloc[258, 34], color='red');
# Neck point
plt.scatter(-key_joint.iloc[258, 101], -key_joint.iloc[258, 102], color='orange');
```

```python
import math
x1 = key_joint['neck_estimate_x']
x2 = key_joint['nose_x']
y1 = key_joint['neck_estimate_y']
y2 = key_joint['nose_y']

nose_neck_angles = np.arctan2(y1-y2, x1-x2) * 180 / np.pi
# print(nose_neck_angles[258])
key_joint['nose_neck_angles'] = nose_neck_angles
plt.hist(nose_neck_angles, bins=20);
```

```python
for i in np.arange(200, 250, 1):
    # Body points
    plt.scatter(key_joint.iloc[i, 1:34], -key_joint.iloc[i, 34:67])
    # Nose point
    plt.scatter(key_joint.iloc[i, 1], -key_joint.iloc[i, 34], color='red');
    # Neck point
    plt.scatter(key_joint.iloc[i, 101], -key_joint.iloc[i, 102], color='peru');
    print(nose_neck_angles[i])
    
    # left point
    plt.scatter(key_joint.loc[i, 'left_wrist_x'], -key_joint.loc[i, 'left_wrist_y'], color='lime');
    
    # right point
    plt.scatter(key_joint.loc[i, 'right_wrist_x'], -key_joint.loc[i, 'right_wrist_y'], color='hotpink');
    plt.xlim(left=0, right=1500)
    plt.ylim(bottom=-900, top=0)
    plt.show();
```

# Updating orientation

(0,0) is in the top left corner of the video frame. The video frame dimensions are 1280 (width) and 720 (height)

The points take into account 

```python
new_names = np.array([])
for col in key_joint.columns:
    if "left" in col:
        new_col = col.replace("left", "right")
    elif "right" in col:
        new_col = col.replace("right", "left")
    else:
        new_col = col
    new_names = np.append(new_names, new_col)
```

```python
switch_lr = key_joint.set_axis(new_names, axis=1, inplace=False)
switch_lr
```

```python
# Body points
i=206
fig, (ax1, ax2) = plt.subplots(2, figsize=[20,20])

# plot 1
ax1.scatter(switch_lr.iloc[i, 1:34], switch_lr.iloc[i, 34:67])
# Nose point
ax1.scatter(switch_lr.iloc[i, 1], switch_lr.iloc[i, 34], color='red');
# Neck point
ax1.scatter(switch_lr.iloc[i, 101], switch_lr.iloc[i, 102], color='peru');
print(nose_neck_angles[i])

# left point
ax1.scatter(switch_lr.loc[i, 'left_wrist_x'], switch_lr.loc[i, 'left_wrist_y'], color='lime');

# right point
ax1.scatter(switch_lr.loc[i, 'right_wrist_x'], switch_lr.loc[i, 'right_wrist_y'], color='hotpink');
ax1.set_xlim(left=0, right=1280)
ax1.set_ylim(bottom=0, top=720)

# plot 2
ax2.scatter(-switch_lr.iloc[i, 1:34], -switch_lr.iloc[i, 34:67])
# Nose point
ax2.scatter(-switch_lr.iloc[i, 1], -switch_lr.iloc[i, 34], color='red');
# Neck point
ax2.scatter(-switch_lr.iloc[i, 101], -switch_lr.iloc[i, 102], color='peru');
print(nose_neck_angles[i])

# left point
ax2.scatter(-switch_lr.loc[i, 'left_wrist_x'], -switch_lr.loc[i, 'left_wrist_y'], color='lime');

# right point
ax2.scatter(-switch_lr.loc[i, 'right_wrist_x'], -switch_lr.loc[i, 'right_wrist_y'], color='hotpink');
ax2.set_xlim(left=-1280, right=0)
ax2.set_ylim(bottom=-720, top=0);
```
