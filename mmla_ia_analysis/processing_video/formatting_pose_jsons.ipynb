```python
import numpy as np
import pandas as pd
```

```python
import json
  
# Opening JSON file
f = open('blazepose_jsons/girl1.json')
  
# returns JSON object as 
# a dictionary
data = json.load(f)

# Closing file
f.close()
```

```python
times = pd.json_normalize(data)['time']

# check 851 rows
len(np.unique(times))
```

```python
df = pd.json_normalize(data[0]['poses'], record_path=['keypoints'])
df.head(3)
```

# set up for iteration

```python
col_names = np.array([])

# add x
x_names = (df['name'].astype(str) + "_x").to_numpy()
# add y 
y_names = (df['name'].astype(str) + "_y").to_numpy()
# add z
z_names = (df['name'].astype(str) + "_z").to_numpy()
# add score
score_names = (df['name'].astype(str) + "_score").to_numpy()

col_names = np.append(col_names, x_names)
col_names = np.append(col_names, y_names)
col_names = np.append(col_names, z_names)
col_names = np.append(col_names, score_names)

print(len(np.unique(col_names)) == len(col_names))
print(len(col_names))
```

```python
df_clean=pd.DataFrame(columns=col_names)
df_clean
```

# Iterate for each frame in video

```python
for i in np.arange(len(data)):
    df_temp = pd.json_normalize(data[i]['poses'], record_path=['keypoints'])
    if len(df_temp)>0:
        x_values = df_temp['x'].tolist()
        y_values = df_temp['y'].tolist()
        z_values = df_temp['z'].tolist()
        score_values = df_temp['score'].tolist()

        all_values = np.append(x_values, y_values)
        all_values = np.append(all_values, z_values)
        all_values = np.append(all_values, score_values)
        len(all_values)

        df_clean.loc[i] = all_values
```

```python
df_final = df_clean.reset_index().rename(columns={"index":'frame'})
df_final
```
